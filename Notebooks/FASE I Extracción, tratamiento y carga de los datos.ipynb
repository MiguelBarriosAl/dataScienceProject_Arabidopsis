# Extract-Transform-Load
Three database functions that are combined into one tool to pull data out of one database and place it into another database. Extract is the process of reading data from a database. In this stage, the data is collected, often from multiple and different types of sources.
In this case I go to extract through a python program divide in two functions: extract function and tranform_load

```python
import numpy as np
import os
import os.path as path
import pandas as pd
import re
import zipfile
```

```python
path='C:/Users/Usuario/Desktop/MBA/TFM/Data'
unzipped='C:/Users/Usuario/Desktop/MBA/TFM/DataEDA/'

def extract(path, unzipped):
    """
        Funtion with two arguments         
        to extract the files with extension
        _tair.txt with regular phrase:
        
            -path of data raw in file.zip
            -path of data unzipped 

        In this file is the expression value
        of the gene=Fold_Change
    """
    patron = re.compile(r'([A-Z]|[a-z]|[0-9]|\_)*\_tair.txt')
    patronGen=re.compile(r'(([A-Z]|[a-z]|[0-9])*\;(.))')
    
    """ Extract Content from File.zip for to save in DataEDA """
    content = os.listdir(path)
    for c in content:
        path = 'C:/Users/Usuario/Desktop/MBA/TFM/Data/'+c+'/'
        pathList = os.listdir(path)
        for p in pathList:
            r = path+p
            file = zipfile.ZipFile(r)
            file.extractall(unzipped)
            contentUnzipped = os.listdir(unzipped)
            for d in contentUnzipped:
                if not patron.match(d):
                    file_1 = unzipped+d
                    os.remove(file_1)
extract(path,unzipped)    
```

```python
unzipped='C:/Users/Usuario/Desktop/MBA/TFM/DataEDA/'

def tranform_load(unzipped):
    """
        To Extract the Data Information
        for to create a Dataframe
    """
    df_raw=pd.DataFrame()
    assays=[]
    contentUnzipped=os.listdir(unzipped)
    """
        Are the relations of the directory where 
        the zip files have been unzipped
    """
    for d in contentUnzipped:
        patron=re.compile(r'([A-Z]|[a-z]|[0-9]|\_)*\_tair.txt')
        rutaD=unzipped
        if patron.match(d):
            """ Split the file name of the extension and
                and to save in array 
            """
            assay=d.split('.')
            assays=np.append(assays, assay[0])
            rutaD= rutaD+d
            df=pd.read_csv(rutaD, usecols=[1, 5], delimiter="\t", header = 0)
            """Clean the NaN and Duplicate values"""
            df=df.dropna()
            df.drop_duplicates(keep='first', inplace=True, subset='LOCUS')
            """Merge in the Dataframe"""
            if not df_raw.empty:
                df=pd.merge(df, df_raw, on='LOCUS', how='inner')
            df_raw=df

    """Rename the list of columns and set index"""
    assays = assays.tolist()
    assays = ['LOCUS'] + assays
    df.columns = assays
    dft = df.set_index('LOCUS').T
    df = df.set_index('LOCUS')
    
    """Save into file.csv"""
    dft.to_csv('C:/Users/Usuario/Desktop/MBA/TFM/DataEDA/AGDAT.csv')
    df.to_csv('C:/Users/Usuario/Desktop/MBA/TFM/DataEDA/AGDA.csv')    
    return df
tranform_load(unzipped)
```

```python
def clean_duplicate():
    cont=0
    df = tranform_load(unzipped)
    dataAntes=df.shape
    patronGen=re.compile(r'(([A-Z]|[a-z]|[0-9])*\;(.))')
    """
        Initialize the variable genes, which will contain 
        the names of all the genes that we have collected 
        in the data matrix.
    """
    genes=[]
    """
        Separate duplicates
    """
    for d in df:
        if patronGen.match(d):
            gen=d.split(';')
            for g in gen:
                genes=np.append(genes, g)
                genDF=pd.DataFrame(df[d])
                genDF=genDF.rename(columns = {d : g})
                df=pd.concat([df, genDF], axis=1, sort=False)
                cont=cont+1
            df.drop(columns=[d], inplace=True)
        else:
            genes=np.append(genes, d)
    df.to_csv('C:/Users/Usuario/Desktop/MBA/TFM/DataEDA/AGDA.csv')
    return df
df = clean_duplicate()
```
